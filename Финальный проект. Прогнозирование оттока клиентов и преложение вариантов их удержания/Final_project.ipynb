# Разработка модели предсказания оттока клиентов телекоммуникационной компании
На основе информации о клиентах оператора связи, построить модель способную прогнозировать отток.

## Описание услуг
Оператор предоставляет два основных типа услуг:

- Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:

- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

## Описание данных

Данные состоят из файлов, полученных из разных источников:

- contract_new.csv — информация о договоре;
- personal_new.csv — персональные данные клиента;
- internet_new.csv — информация об интернет-услугах;
- phone_new.csv — информация об услугах телефонии.
- Во всех файлах столбец customerID содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

## Подготовка данных к обучению

Разделим выборку на обучающую и тестовую

````python
features = models_data.drop('target', axis = 1)
target = models_data['target']
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size = 0.25, random_state = RANDOM_STATE, shuffle = True, stratify = target)

print(features_train.shape, target_train.shape)
print(features_test.shape, target_test.shape)
````

Зададим переменные для скалирования непрерывных признаков и кодирования категориальных признаков

````python
scaler = StandardScaler()
ohe = OneHotEncoder(drop='first', handle_unknown='error')
oe = OrdinalEncoder(handle_unknown='error')

scaler_columns = ['monthly_charges', 'contract_duration']

ohe_columns = features.columns.to_list()
ohe_columns.remove('monthly_charges')
ohe_columns.remove('contract_duration')

oe_columns = features.columns.to_list()
oe_columns.remove('monthly_charges')
oe_columns.remove('contract_duration')
````

Напишем функцию для трансформации признаков

````python
col_transformer = make_column_transformer(
        (ohe, ohe_columns),
        (scaler, scaler_columns),
        remainder='passthrough',
        verbose_feature_names_out=False)
````

Напишем функцию для трансформации признаков "деревянных" моделей

````python
col_transformer_tree = make_column_transformer(
        (oe, oe_columns),
        (scaler, scaler_columns),
        remainder='passthrough',
        verbose_feature_names_out=False)
````

## Обучение моделей

Для обучения были выбраны три модели от простой к сложной. Это логистическая регрессия, решающее дерево и категориальный бустинг. Перейдем к их обучению и сравнению качества на кросс-валидации

````python
model_1 = LogisticRegression(random_state = RANDOM_STATE)
pipeline_1 = Pipeline([('col', col_transformer), ('lr', model_1)])
params_1 = {'lr__C': np.arange(1, 11, 2)}

search_1 = GridSearchCV(pipeline_1, params_1, cv = 5, scoring = 'roc_auc')
````

<div class="alert alert-block alert-success">

````python
print('Лучшие гиперпараметры: ', model_1.best_params_)
print('Лучшее значение ROC-AUC на валидационной выборке: ', model_1.best_score_)
````

````python
model_2 = DecisionTreeClassifier(random_state = RANDOM_STATE)
pipeline_2 = Pipeline([('col', col_transformer_tree), ('dtc', model_2)])
params_2 = {'dtc__max_depth': [None] + [i for i in range(2, 9)],
              'dtc__min_samples_split': np.arange(2,5,1),
              'dtc__min_samples_leaf': np.arange(1,5,1)}

search_2 = GridSearchCV(pipeline_2, params_2, cv = 5, scoring = 'roc_auc')
````
